# Exploratory Analysis:

In this section we will explore the delay causes of the airline and will able to know at the end that which is the main cause of delay of flight we will easily interpret.

### Defining function for labeling axes and title

```python
def ad_label(title,x,y):
    plt.title(title,size=18)
    plt.xlabel(x,size=15)
    plt.ylabel(y,size=15)
```

## 1)Uni-Variate Analysis:

### Which Airlines is mostly used in the season 2019 by the traveller?

### Defining function for univariate count plot

```python

def scplot(name):
    df_clean=pd.read_csv("clean_flight.csv")
    #base_color = sb.color_palette()[0]
    orders= df_clean[name].value_counts().sort_values(ascending = False).index
    
    plt.figure(figsize=[8,6])
    sb.countplot(data = df_clean,y=name,order = orders )
    
    
```

```python
scplot('carrier_name')
ad_label("Most Running Flights","Count","Carrier Name")
max_c=df_clean['carrier_name'].value_counts().max()
max_pro=max_c/df_clean.shape[0]
tick_lab = np.arange(0,max_pro,0.02)
tick_val=['{:0.2f}'.format(v) for v in tick_lab]
df2= df_clean['carrier_name'].value_counts()
for i in range(df2.shape[0]):
        count = df2[i]
        pcd_string = "{:0.2f}%".format(count*100/df_clean.shape[0])
        plt.text(count+1,i,pcd_string)
```

**OBSERVATIONS:** SkyWest Airlines Inc. were the most popular airlines used by the people for travelling.

### which Month was the bussiest month i.e more airlines were flyed?

## BI-VARIATE :

### Defining function for bivariate reg plot

```python
def bivar(name1,name2):
    plt.figure(figsize=[8,6])
    sb.regplot(data=df_clean,x=name1,y=name2,x_jitter=0.1,y_jitter=0.1,scatter_kws={'alpha':1/3})
```

## Analyse the relationship between the arrival delay and the delay causes.(Top #3)

```python
bivar('arr_del15','weather_ct1')
ad_label("NO. OF ARRIVAL DELAYS DUE TO WEATHER","ARRIVAL DELAY COUNT","WEATHER DELAY COUNT")
```

```python
bivar('arr_del15','carrier_ct')
ad_label("NO. OF ARRIVAL DELAYS DUE TO CARRIER","ARRIVAL DELAY COUNT","CARRIER DELAY COUNT")
```

```python
bivar('arr_del15','late_aircraft_delay')
ad_label("NO. OF ARRIVAL DELAYS DUE TO LATE AIRCRAFT","ARRIVAL DELAY COUNT","CARRIER DELAY COUNT")
```

#### Observation : the arrival delay and the other 3 delay causes were found to have positive correlation between them and after visualizing the Bivariate graph it is seen thaat late aircraft delay cause was having the highest correlation when compared to other two.
